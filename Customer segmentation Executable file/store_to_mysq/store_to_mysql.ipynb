```python
import os
import datetime
import pandas as pd
from sqlalchemy import create_engine
import pymysql

start_time = datetime.datetime.now()
print('Begin:', start_time)
num = 0

engine = create_engine('mysql+pymysql://root:123@localhost:3306/customer?charset=utf8mb4')
path = r'C:\Users\Admin\OneDrive\Documents\Major_project\Final_project\data'
files = os.listdir(path)

for i in files:
    data = pd.read_csv(os.path.join(path, i), header=0)
    data.to_sql(name='world_trade', con=engine, index=False, if_exists='append')
    num += 1
    print('Imported: ', i)

end_time = datetime.datetime.now()
print(end_time)

total_time = end_time-start_time
print('Totaltime: ', total_time)

print("Total no. of files: ", num)
```
