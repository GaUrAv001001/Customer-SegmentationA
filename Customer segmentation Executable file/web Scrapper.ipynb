```python

import pandas as pd
import requests
from bs4 import BeautifulSoup
Customer_name = []
places = []
ratings = []
reviews = []
discription = []
rt_class = ["_3LWZlK _1BLPMq","_3LWZlK _1rdVr6 _1BLPMq","_3LWZlK _32lA32 _1BLPMq"]
rev_class = ["","_3LWZlK _1BLPMq"]
```

```python
for i in range(1,3):
    url = "https://www.flipkart.com/samsung-galaxy-f04-opal-green-64-gb/product-reviews/itmd0c05f8b03876?pid=MOBGKY2V7QZ2QVFS&lid=LSTMOBGKY2V7QZ2QVFSGOGYVU&marketplace=FLIPKART&page="+str(i)
    r = requests.get(url)
    # print(r)

    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div", class_="_1YokD2 _3Mn1Gg col-9-12")
    # box = soup.find("div", class_="_6t1WkM _3HqJxg")

    # Customer Name
    names = box.find_all("p", class_ = "_2sc7ZR _2V5EHH")
    for i in names:
        name = i.text
        Customer_name.append(name)

    # Customer Review
    rev = box.find_all("p", class_="_2-N8zT")
    for i in rev:
        review = i.text
        reviews.append(review)

    # Customer Place
    plc = box.find_all("p", class_="_2mcZGG")
    for i in plc:
        place = i.text
        places.append(place)



    rt = box.find_all("div", class_ = rt_class)
    for i in rt:
        rating = i.text
        ratings.append(rating)



    disc = box.find_all("div", class_="t-ZTKy" )
    for i in disc:
        Desc = i.text
        discription.append(Desc)




df = pd.DataFrame({"Customer Name":Customer_name,"Reviews":reviews,"Place":places,"Ratings":ratings,"Description":discription})
print(df)
# print(len(Customer_name)//10)



df.to_csv("Flipkart_customer_review_dataA.csv")
# print(soup)

    # nextPage = soup.find("a", class_ = "ge-49M").get("href")
    # cnp = "https://www.flipkart.com"+nextPage
    # print(cnp)

# url = cnp
# r = requests.get(url)
# soup = BeautifulSoup(r.text,"lxml")
```
